```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
img='copy.jpg'
```

```python
reader=easyocr.Reader(['en'],gpu=False)
result=reader.readtext(img)
result
```

```python
top_left=tuple(result[0][0][0])
bottom_right=tuple(result[0][0][2])
text=result[0][1]
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
img1=cv2.imread(img)
img1=cv2.rectangle(img1,top_left,bottom_right,(0,255,0),5)
img1=cv2.putText(img1,text,top_left,font,.5,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img1)
plt.show()
```

```python
img1=cv2.imread(img)
for detection in result:
  top_left=tuple([int(val) for val in detection[0][0]])
  bottom_right=tuple([int(val) for val in detection[0][2]])
  text=detection[1]
  font=cv2.FONT_HERSHEY_SIMPLEX
  img1=cv2.rectangle(img1,top_left,bottom_right,(0,255,0),5)
  img1=cv2.putText(img1,text,top_left,font,4,(255,255,255),2,cv2.LINE_AA)
plt.figure(figsize=(10,10))
plt.imshow(img1)
plt.show()
```
